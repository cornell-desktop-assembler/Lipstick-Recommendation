```python
vocab = set()

for item in TF_SENTIMENT.values():
    for word in item.keys():
        vocab.add(word)
```

```python
len(vocab)
```

```python
with open("google-10000-english.txt", "r", encoding="utf-8") as fin:
    g10k = fin.read().split()
```

```python
len(g10k), g10k[:10]
```

```python
g10k = set(g10k)
```

```python
vocab_comb = set.intersection(g10k, vocab)
```

```python
vocab_comb = [w for w in vocab_comb if len(w) >= 3]
```

```python
len(vocab_comb)
```

```python
sorted(vocab_comb)
```

```python
import json

with open("vocab_comb.json", "w", encoding="utf-8") as fout:
    json.dump(vocab_comb, fout)
```

```python
with open("listing.txt") as fin:
    t = fin.read()
```

```python
import requests
```

```python
import time
```

```python
import json
```

```python
import re
```

```python
pids = set(re.findall(r"(?:grid:)(.*?)(?::)", t))
pids.add("P443336")
```

```python
len(pids)
```

```python
URL_TEMPLATE = \
    "https://api.bazaarvoice.com/data/reviews.json?"\
    "Filter=ProductId%3A{pid}"\
    "&Limit=100&Offset={offset}"\
    "&Include=Products%2CComments"\
    "&Stats=Reviews"\
    "&passkey=rwbw526r2e7spptqd2qzbkp7&apiversion=5.4"
```

```python
for i, pid in enumerate(pids):
    def scrape_and_save(pid, offset):
        print(offset, end="")
        url = URL_TEMPLATE.format(pid=pid, offset=offset)
        response = requests.get(url)
        text = response.text
        with open("raw jsons/{pid}_{offset}.json".format(pid=pid, offset=offset), "w", encoding="utf-8") as f:
            f.write(text)
        print(".", end="")
        time.sleep(1)
        return text
    
    offset=0
    print("{i:>4d}/{t} {pid}_{offset}... ".format(i=i, t=len(pids), pid=pid, offset=offset), end="")

    text = scrape_and_save(pid, offset*100)

    j = json.loads(text)
    total = j["TotalResults"]
    print("<total={}>".format(total), end="")
    max_off = int(total/100)

    for offset in range(1, max_off+1):
        scrape_and_save(pid, offset*100)
    
    print("\ndone")
```

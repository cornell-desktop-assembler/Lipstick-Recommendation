```python
import os
import json
data_path = os.path.join(
    "sku2desc.json"
)

with open(data_path, encoding="utf-8") as fin:
    sku2desc = json.load(fin)
```

```python
sku2desc
```

```python
brand_set = set()
for sku, desc in sku2desc.items():
    brand = desc["brand"]
    brand_set.add(brand)
```
